```
X = df[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary' ]]
Y = df['Exited']
```

```
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,Y)
```

```
import statsmodels.api as sm
X = sm.add_constant(X)
lm_1 = sm.OLS(Y,X).fit()
print(lm_1.summary())
```
